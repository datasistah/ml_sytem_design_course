```python
#from peft import PeftModel
from transformers import LLaMATokenizer, LLaMAForCausalLM, GenerationConfig, pipeline
import textwrap
```

```python
from langchain import PromptTemplate, LLMChain
```

```python

tokenizer = LLaMATokenizer.from_pretrained("decapoda-research/llama-7b-hf")

model = LLaMAForCausalLM.from_pretrained(
    "decapoda-research/llama-7b-hf",
    load_in_8bit=True,
    device_map="auto",
)
#model = PeftModel.from_pretrained(model, "samwit/alpaca7B-lora")
```

```python
from langchain.llms.huggingface_pipeline import HuggingFacePipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    temperature=0.6,
    top_p=0.95,
    repetition_penalty=1.2,
    max_new_tokens=256

)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
template = '''I want you to act as a scrum manager of a company.  Write user stories in the following format

      Description: As a user, I want to be able to recover the password to my account, so that I will be able to access my account in case I forgot the password.
      Acceptance Criteria:
      Given the user receives the link via the email
      When the user navigates through the link received in the email
      Then the system enables the user to set a new password

      Description: As a data scientist, I want to be able to collect the results of sending our gold set documents through the current API, so that these results can be used to establish baseline evaluation metrics for the API.
      Acceptance Criteria:
      Given the data scientist has a gold set of documents to be sent through the API.
      When the data scientist sends the gold set documents through the API.
      Then the API should return results that can be collected and used to establish baseline evaluation metrics for the API.

      Description:{description}
      Acceptance Criteria:

      '''
prompt = PromptTemplate.from_template(template)
#prompt = PromptTemplate(template=template, input_variables = ['description'])

llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
description = "As a data scientist, I want to evaluate the performance of the model with the updated changes, so that we have a clear understanding in the change in performance due to this update."

#llm_chain.run(description)
```

```python
#https://www.pinecone.io/learn/langchain-prompt-templates/
from langchain import FewShotPromptTemplate

examples = [
    {
    "description": "As a user, I want to be able to recover the password to my account, so that I will be able to access my account in case I forgot the password.",
    "criteria":"GIVEN the user receives the link via the email,\
    WHEN the user navigates through the link received in the email,\
    THEN the system enables the user to set a new password."
    },

    {
      "description": "As a data scientist, I want to be able to collect the results of sending our gold set documents through the current API, so that these results can be used to establish baseline evaluation metrics for the API.",
      "criteria":"GIVEN the data scientist has a gold set of documents to be sent through the API.\
      WHEN the data scientist sends the gold set documents through the API.\
      THEN the API should return results that can be collected and used to establish baseline evaluation metrics for the API."
      }
]
example_formatter_template = """
Description: {description}
Acceptance Criteria: {criteria}
"""

example_prompt = PromptTemplate(
    input_variables=["description", "criteria"],
    template=example_formatter_template
    )
# the prefix is our instructions
prefix = """
You are a scrum manager of a team and need to write the acceptance criteria for user stories based on the description. The following are examples of user stories:
"""
suffix = """
Description: {description}
Acceptance Criteria:
"""
# Finally, we create the `FewShotPromptTemplate` object.
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix=prefix,
    suffix=suffix,
    input_variables=['description'],
    example_separator="\n\n"

)
```

```python
example="As a data scientist, I want to evaluate the performance of the model with the updated changes, so that we have a clear understanding in the change in performance due to this update."
print(few_shot_prompt.format(description=example))
```

```python
llm_chain = LLMChain(prompt=few_shot_prompt, llm=llm)
```

```python
print(llm_chain.run(example))
```

```python
import gradio as gr

def ado_acceptance_criteria(text):
  response_list = []
  response_dict = {}
  response = llm_chain.run(text)
  portion_before_quotes = response.split('"""')[0]
  #response_dict['Description'] =text
  formatted_acceptance_criteria = portion_before_quotes.replace("Given", "\nGiven").replace("When", "\nWhen").replace("Then", "\nThen")
  response_dict['Acceptance Criteria'] = portion_before_quotes
  #response_dict['Acceptance Criteria'] = formatted_acceptance_criteria
  response_list.append(response_dict)
  return response_list

ado_story = gr.Interface(
    fn=ado_acceptance_criteria,
    description="Created by Tiffany Teasley",
    inputs= gr.Textbox(label='Enter story desciption', placeholder='As a [type of user], I want to [perform some task] so that I can [achieve some goal'),
    # examples=[
    #      ["As a [type of user], I want to [perform some task] so that I can [achieve some goal]." ],
    #      ["Acceptance Criteria", "GIVEN that [some context], WHEN [some action is carried out], THEN [a set of observable outcomes should occur]."]
    # ],
    title="Agile Story Helper",
    outputs="json",
    theme='gstaff/sketch')

ado_story.launch(share=True)
```
